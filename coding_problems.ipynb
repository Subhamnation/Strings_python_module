### Count Vowels and Consonants

#### 1.  Assume all characters are in lowercase 
#### 2. String may include special characters 

##### My approach -- Wrong

```python
string_1 = '__hel lo 123$world '
vowels = ['a','e','i','o','u']
count = 0
for i in string_1:
    if i in vowels:
        count = count +1
print(f"The number of vowels: {count}")
print(f"The number of consonants and special character: {len(string_1) - count}")
```

#### Instructor approach --- Which is the correct only

```python
s = '__hel lo 123$world '

vowels = 0
consonants = 0
for ch in s:
    if ch.isalpha():
        #print(ch)
        if ch in 'aeiou':
            vowels += 1
        else:
            consonants += 1
print(f"The number of vowels are: {vowels} & consonants are {consonants}")
```

#### Frequency of Characters

```python
s1 = 'banana'
## Expected outcome: b = 1, a = 3, n =2
```

##### Dictionary module

```python
counts = dict()
for ch in s1:
    counts[ch] = counts.get(ch,0) + 1
print(counts)
```

```python
s2 = 'malayalam'
counts = dict()
for ch in s2:
    counts[ch] = counts.get(ch,0) + 1
print(counts)
```
