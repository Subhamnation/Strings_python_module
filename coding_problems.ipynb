#### Instructor approach --- Which is the correct only

```python
s = '__hel lo 123$world '

vowels = 0
consonants = 0
for ch in s:
    if ch.isalpha():
        #print(ch)
        if ch in 'aeiou':
            vowels += 1
        else:
            consonants += 1
print(f"The number of vowels are: {vowels} & consonants are {consonants}")
```

#### Frequency of Characters

```python
s1 = 'banana'
## Expected outcome: b = 1, a = 3, n =2
```

##### Dictionary module

```python
counts = dict()
for ch in s1:
    counts[ch] = counts.get(ch,0) + 1
print(counts)
```

```python
s2 = 'malayalam'
counts = dict()
for ch in s2:
    counts[ch] = counts.get(ch,0) + 1
print(counts)
```

#### Anagrams

* Return True if two strings contain the same characters in any order 
* Strings will be of same length and lowercase 

```python
s4 = 'listen'
s5 = 'silent'
```

##### First approach
* Sorting approach [Time: nlog(n)]

```python
s5 = sorted(s5)
print(s5)
s4 = sorted(s4)
print(s4)
if s4 == s5:
    print("Anagram")
else:
    print("Not Anagram")
```

* Approach 2

```python
s4 = 'listen'
s5 = 'silent'
```

```python
mapping = [0] * 26
mapping
```

```python
for ch in s4:
    index = ord(ch) - 97
    mapping[index] += 1
for ch in s5:
    index = ord(ch) - 97
    mapping[index] -= 1
mapping
if all(m == 0 for m in mapping):
    print("Anagram")
else:
    print("Not anagram")
```
