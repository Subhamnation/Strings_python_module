### Strings demo


#### Strings in python 

```python
str1 = "This is a string in python"
print(str1)
print(type(str1))
```

```python
str2 = 'This is a string in python'
print(str2)
print(type(str2))
```

```python
str3 = """This is a string in python"""
print(str3)
print(type(str3))
```

```python
str4 = '''This is a string in python'''
print(str4)
print(type(str4))
```

#### Properties of string

##### Immutable

```python
str5 = "Hello"
print(id(str5)) # memory address of the object where ot os stored 
```

```python
str5 = str5 + ' World'
print(str5)
print(id(str5))
```

##### Iterable

```python
str5
```

```python
for ch in str5:
    print(ch)
```

```python
for ch in str5:
    print(ch, end ="")
```

#### Unicode

```python
str5
```

```python
ord('a'), ord('z')
```

```python
ord('A'), ord('Z')
```

```python
ord('&'), ord('*'), ord('$')
```

```python
### Note that, all characters in python are represented by their corresponding ASCII values. So, when we use the ord() function, it returns the ASCII value of the character.

for ch in str5:
    print(ch, ord(ch))
```
